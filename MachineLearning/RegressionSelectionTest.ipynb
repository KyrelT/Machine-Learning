```python
import sklearn
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
```

```python
x, y = make_regression(n_samples = 50, n_features = 1, n_informative = 1)
```

```python
sns.regplot(x=x,y=y)
```

```python
a,b = make_regression(n_samples = 50, n_features = 6, n_targets = 1)
```

```python
fs = SelectKBest(score_func = f_regression, k=6)
```

```python
X_selected = fs.fit_transform(a,1)
```

```python
print(X_selected.shape)
```

```python
plt.plot(X_selected,b,'o')
```
